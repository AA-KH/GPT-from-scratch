```python
with open('input.txt', 'r') as f:
    text = f.read()
```

```python
print(len(text))
chars = sorted(list(set(text)))
print(len(chars))
```

```python
#TOKENIZATION
stoi = {ch:i for i,ch in enumerate(chars)} #string to integer conversion
#enumerate(chars) gives (0,a), (1,b), (2,c)
itos = {i:ch for i, ch in enumerate(chars)} #integer to string

def encode(s):
    return [stoi[c] for c in s]

def decode(i):
    return ''.join([itos[n] for n in i])

print(encode("car"))
print(decode(encode("car")))
```

```python
import torch
data = torch.tensor(encode(text), dtype=torch.long)
```

```python
#SPLITTING INTO TRAINING AND VALIDATION DATASET
n = int(0.9*len(data))
train_data = data[:n]
val_data = data[n:  ]
```
